# Join:

### 1. Provide a table for all the for all web_events associated with the account name of Walmart. Include the primary_poc, time of the event, and the channel for each event. 

```python
%%sql 
SELECT a.primary_poc, w.occurred_at, w.channel, a.name
FROM web_events w
JOIN accounts a
ON w.account_id = a.id
WHERE a.name = 'Walmart';
```

### 2. Provide a table that provides the region for each sales_rep along with their associated accounts.  
### Include three columns: the region name, the sales rep name, and the account name. 
### Sort account name.


```python
%%sql
SELECT r.name region, s.name rep, a.name account
FROM sales_reps s
JOIN region r
ON s.region_id = r.id
JOIN accounts a
ON a.sales_rep_id = s.id
ORDER BY a.name;
```

### 3. Provide the name for each region for every order, as well as the account name and the unit price they paid (total_amt_usd/total) for the order. 


```python
%%sql
SELECT r.name region, a.name account, 
 o.total_amt_usd/(o.total + 0.01) unit_price
FROM region r
JOIN sales_reps s
ON s.region_id = r.id
JOIN accounts a
ON a.sales_rep_id = s.id
JOIN orders o
ON o.account_id = a.id;
```

### 4. Provide a table that provides the region for each sales_rep along with their associated accounts in Midwest.
### Include the region name, the sales rep name, and the account name.
### Sort the account name.

```python
%%sql
SELECT r.name region, s.name rep, a.name account
FROM sales_reps s
JOIN region r
ON s.region_id = r.id
JOIN accounts a
ON a.sales_rep_id = s.id
WHERE r.name = 'Midwest'
ORDER BY a.name;
```

### 5. Provide a table that provides the region for each sales_rep along with their associated accounts.
### Look for accounts where the sales rep has a first name starting with S and in the Midwest region.
### Include the region name, the sales rep name, and the account name.
### Sort the account name.


```python
%%sql
SELECT r.name region, s.name rep, a.name account
FROM sales_reps s
JOIN region r
ON s.region_id = r.id
JOIN accounts a
ON a.sales_rep_id = s.id
WHERE r.name = 'Midwest' AND s.name LIKE 'S%'
ORDER BY a.name;
```

### 6. Provide a table that provides the region for each sales_rep along with their associated accounts. 
### Look for accounts where the sales rep has a last name starting with K and in the Midwest region. 


```python
%%sql
SELECT r.name region, s.name rep, a.name account
FROM sales_reps s
JOIN region r
ON s.region_id = r.id
JOIN accounts a
ON a.sales_rep_id = s.id
WHERE r.name = 'Midwest' AND s.name LIKE '% K%'
ORDER BY a.name;
```

### 7. Provide the name for each region for every order, as well as the account name and the unit price they paid  (total_amt_usd/total) for the order. 
### Look for accounts where standard order quantity exceeds 100.


```python
%%sql
SELECT r.name region, a.name account, o.total_amt_usd/(o.total + 0.01) unit_price
FROM region r
JOIN sales_reps s
ON s.region_id = r.id
JOIN accounts a
ON a.sales_rep_id = s.id
JOIN orders o
ON o.account_id = a.id
WHERE o.standard_qty > 100;
```

### 7. Provide the name for each region for every order, as well as the account name and the unit price they paid  (total_amt_usd/total) for the order.
### Include account where standard order quantity exceeds 100 and the poster order quantity exceeds 50.


```python
%%sql
SELECT r.name region, a.name account, o.total_amt_usd/(o.total + 0.01) unit_price
FROM region r
JOIN sales_reps s
ON s.region_id = r.id
JOIN accounts a
ON a.sales_rep_id = s.id
JOIN orders o
ON o.account_id = a.id
WHERE o.standard_qty > 100 AND o.poster_qty > 50
ORDER BY unit_price;
```

### 9. Provide the name for each region for every order, as well as the account name and the unit price they paid  (total_amt_usd/total) for the order. 
### Include accounts where standard order quantity exceeds 100 and the poster order quantity exceeds 50. 


```python
%%sql
SELECT r.name region, a.name account, o.total_amt_usd/(o.total + 0.01) unit_price
FROM region r
JOIN sales_reps s
ON s.region_id = r.id
JOIN accounts a
ON a.sales_rep_id = s.id
JOIN orders o
ON o.account_id = a.id
WHERE o.standard_qty > 100 AND o.poster_qty > 50
ORDER BY unit_price DESC;
```

### 10. What are the different channels used by account id 1001?


```python
%%sql
SELECT DISTINCT a.name, w.channel
FROM accounts a
JOIN web_events w
ON a.id = w.account_id
WHERE a.id = '1001';
```

### 11. Find all the orders that occurred in 2015


```python
%%sql
SELECT o.occurred_at, a.name, o.total, o.total_amt_usd
FROM accounts a
JOIN orders o
ON o.account_id = a.id
WHERE o.occurred_at BETWEEN '01-01-2015' AND '01-01-2016'
ORDER BY o.occurred_at DESC;
```
